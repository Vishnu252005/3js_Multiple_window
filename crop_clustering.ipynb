# Crop Pattern Clustering for Sustainable Farming - Comprehensive Report

This notebook is focused on **Crop Pattern Clustering for Sustainable Farming**. It uses clustering techniques on the FAO Crop Production Dataset to identify different crop patterns based on 'Area harvested' and 'Yield'.

The main algorithms used are:

*   **K-Means Clustering:** This algorithm groups data points into a predefined number of clusters based on how close they are to the center of each cluster.
*   **Expectation-Maximization (EM) Clustering (using Gaussian Mixture Models):** This is a probabilistic approach that assumes the data comes from a mix of different probability distributions and assigns data points to clusters based on the likelihood they belong to each distribution.

The notebook also uses several metrics to evaluate how well these algorithms are working, such as Inertia, Silhouette Score, BIC, and AIC.

## Objective

The primary objective of this project is to analyze crop patterns using clustering techniques applied to the FAO Crop Production Dataset. This analysis aims to identify distinct patterns based on key features like 'Area harvested' and 'Yield', compare the performance of K-Means and EM clustering algorithms, interpret the ecological meaning of the identified clusters, and discuss the ethical implications of providing access to agri-analytics tools for small farmers. The project aligns with SDG 2 (Zero Hunger) and SDG 12 (Responsible Consumption).

## Theory

This project utilizes unsupervised learning techniques, specifically **K-Means clustering** and **Expectation-Maximization (EM) clustering**.

*   **K-Means Clustering:** An iterative algorithm that partitions data points into a predefined number of clusters (k) based on their proximity to the cluster centroids. It aims to minimize the within-cluster sum of squares.
*   **Expectation-Maximization (EM) Clustering (using Gaussian Mixture Models):** A probabilistic model that assumes data points are generated from a mixture of Gaussian distributions. It iteratively estimates the parameters of these distributions and assigns data points to clusters based on the probability of belonging to each distribution.

Metrics used for evaluating clustering performance include **Inertia** (for K-Means, measures the sum of squared distances of samples to their closest cluster center) and **Silhouette Score** (measures how similar a data point is to its own cluster compared to other clusters). For EM, **Bayesian Information Criterion (BIC)** and **Akaike Information Criterion (AIC)** are used to select the number of components (clusters).

## Dataset

The dataset used is the **FAO Production Crops and Livestock Dataset** from Kaggle. It contains information on various crops and livestock across different regions and years, including metrics like Area harvested, Yield, and Production.

## Preprocessing

The data preprocessing steps involved:
1.  **Loading the dataset:** Reading the `Production_Crops_Livestock_E_All_Data_NOFLAG.csv` file into a pandas DataFrame.
2.  **Selecting relevant columns:** Focusing on 'Area', 'Item', 'Element', and the year-wise columns ('Y1961' to 'Y2023').
3.  **Filtering data:** Selecting rows where 'Element' is either 'Area harvested' or 'Yield'.
4.  **Melting and Pivoting:** Reshaping the data from a wide format (years as columns) to a long format, and then pivoting to have 'Area harvested' and 'Yield' as separate columns for each 'Area', 'Item', and 'Year' combination.
5.  **Handling missing values:** Dropping rows with any missing values after pivoting.
6.  **Sampling:** Taking a 1% sample of the cleaned data (`df_cleaned.sample(frac=0.01, random_state=42)`) to manage the dataset size for computational efficiency.

## Model

**Feature Scaling:**
The selected features ('Area harvested' and 'Yield') were scaled using `StandardScaler` to standardize their range and prevent features with larger values from dominating the clustering process.

**K-Means Clustering:**
The KMeans algorithm was applied to the scaled sampled features. The optimal number of clusters was explored by iterating through a range (2 to 10) and calculating inertia and silhouette scores. For the final evaluation and interpretation, **3 clusters** were chosen.

**EM Clustering:**
The GaussianMixture model (for EM clustering) was applied to the scaled sampled features. The optimal number of components was explored by iterating through a range (2 to 10) and analyzing BIC and AIC scores. For the final evaluation and interpretation, **3 components** were chosen.

## Results

The clustering algorithms partitioned the sampled data points into 3 clusters based on the scaled 'Area harvested' and 'Yield' values.

## Comparative Metrics

The clustering results were evaluated using the Silhouette Score:

*   **Silhouette Score for EM (with 3 components):** -0.14936839110422204

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Production_Crops_Livestock_E_All_Data_NOFLAG.csv', encoding='latin-1')

# Display the first 5 rows
display(df.head())

# Display column names and their data types
display(df.info())

# Display basic statistical summary
display(df.describe())
```

# Data Preprocessing and Sampling

```python
import pandas as pd

# Select relevant columns
year_columns = [col for col in df.columns if col.startswith('Y')]
selected_cols = ['Area', 'Item', 'Element'] + year_columns
df_selected = df[selected_cols]

# Filter for 'Yield' and 'Area harvested'
df_filtered = df_selected[df_selected['Element'].isin(['Yield', 'Area harvested'])]

# Melt the DataFrame
df_melted = df_filtered.melt(id_vars=['Area', 'Item', 'Element'],
                             value_vars=year_columns,
                             var_name='Year',
                             value_name='Value')

# Pivot the melted DataFrame
df_pivoted = df_melted.pivot_table(index=['Area', 'Item', 'Year'],
                                   columns='Element',
                                   values='Value').reset_index()

# Handle missing values by dropping rows with NaNs
df_cleaned = df_pivoted.dropna()

# Sample a smaller subset of the data
df_sampled = df_cleaned.sample(frac=0.01, random_state=42) # Sample 1% of the data

# Display the cleaned and sampled DataFrame
display(df_sampled.head())
display(df_sampled.info())
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

# Select the features for clustering
features = df_cleaned[['Area harvested', 'Yield']]

# Instantiate StandardScaler
scaler = StandardScaler()

# Fit and transform the features
scaled_features = scaler.fit_transform(features)

# Convert the scaled features back to a DataFrame for better readability
scaled_features_df = pd.DataFrame(scaled_features, columns=['Area harvested', 'Yield'])

# Display the first few rows of the scaled features and their shape
display(scaled_features_df.head())
display(scaled_features_df.shape)
```

# Apply K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import pandas as pd

# Make sure scaled_features_df is available from the previous step
# If not, re-run the feature scaling step with the sampled data
# Select the features for clustering from the sampled data
features_sampled = df_sampled[['Area harvested', 'Yield']]

# Instantiate StandardScaler (if not already instantiated)
# Assuming scaler is already instantiated in cell 243341c6
# If not, uncomment and run the following lines:
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()

# Fit and transform the sampled features
scaled_features_sampled = scaler.fit_transform(features_sampled)

# Determine a range of possible cluster numbers
range_n_clusters = range(2, 11)  # Start from 2 clusters for silhouette score

# Initialize lists to store inertia and silhouette scores
inertia = []
silhouette_scores = []

# Iterate through the range of cluster numbers
for n_clusters in range_n_clusters:
    # Instantiate KMeans model
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10) # Explicitly set n_init

    # Fit the KMeans model to the scaled features (using the sampled data)
    kmeans.fit(scaled_features_sampled)

    # Calculate inertia
    inertia.append(kmeans.inertia_)

    # Calculate silhouette score using the scaled sampled features
    score = silhouette_score(scaled_features_sampled, kmeans.labels_)
    silhouette_scores.append(score)

# Plot the Elbow Method (Inertia vs. Number of Clusters)
plt.figure(figsize=(10, 5))
plt.plot(range_n_clusters, inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# Plot the Silhouette Scores vs. Number of Clusters
plt.figure(figsize=(10, 5))
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

# Apply EM Clustering

```python
from sklearn.mixture import GaussianMixture
import numpy as np

# Determine a range of possible cluster numbers
range_n_components = range(2, 11)

# Initialize lists to store BIC and AIC scores
bic_scores = []
aic_scores = []

# Iterate through the range of cluster numbers
for n_components in range_n_components:
    # Instantiate GaussianMixture model
    gm = GaussianMixture(n_components=n_components, random_state=42)

    # Fit the model to the scaled features (using the sampled data)
    gm.fit(scaled_features_sampled)

    # Calculate BIC and AIC
    bic_scores.append(gm.bic(scaled_features_sampled))
    aic_scores.append(gm.aic(scaled_features_sampled))

# Plot the BIC and AIC scores
plt.figure(figsize=(10, 5))
plt.plot(range_n_components, bic_scores, marker='o', label='BIC')
plt.plot(range_n_components, aic_scores, marker='o', label='AIC')
plt.title('BIC and AIC Scores for EM Clustering')
plt.xlabel('Number of Components (Clusters)')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```

# Evaluate Clustering Results

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score

# Assuming optimal_k for KMeans and optimal_n_components for EM are determined from previous steps
# Replace 'None' with the optimal number of clusters/components you determined
optimal_k = 3 # e.g., 3 based on elbow/silhouette plots
optimal_n_components = 3 # e.g., 3 based on BIC/AIC plots

if optimal_k is not None:
    # Apply KMeans with the optimal number of clusters
    kmeans_optimal = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
    kmeans_optimal.fit(scaled_features_sampled)
    kmeans_silhouette = silhouette_score(scaled_features_sampled, kmeans_optimal.labels_)
    print(f"Silhouette Score for KMeans with {optimal_k} clusters: {kmeans_silhouette}")
else:
    print("Optimal number of clusters for KMeans is not determined. Please set 'optimal_k'.")


if optimal_n_components is not None:
    # Apply EM with the optimal number of components
    gm_optimal = GaussianMixture(n_components=optimal_n_components, random_state=42)
    gm_optimal.fit(scaled_features_sampled)
    # For GaussianMixture, we use the predicted cluster labels for silhouette score
    em_labels = gm_optimal.predict(scaled_features_sampled)
    em_silhouette = silhouette_score(scaled_features_sampled, em_labels)
    print(f"Silhouette Score for EM with {optimal_n_components} components: {em_silhouette}")
else:
     print("Optimal number of components for EM is not determined. Please set 'optimal_n_components'.")
```

# Interpret Ecological Meaning of Clusters

```python
# Add KMeans cluster labels to the sampled DataFrame
df_sampled['KMeans_Cluster'] = kmeans_optimal.labels_

# Add EM cluster labels to the sampled DataFrame
df_sampled['EM_Cluster'] = em_labels

# Analyze KMeans clusters
print("K-Means Cluster Interpretation:")
display(df_sampled.groupby('KMeans_Cluster')[['Area harvested', 'Yield']].mean())

# Analyze EM clusters
print("\nEM Cluster Interpretation:")
display(df_sampled.groupby('EM_Cluster')[['Area harvested', 'Yield']].mean())

# You can further analyze the distribution of 'Area' and 'Item' within each cluster
# to get more specific ecological insights.
# For example:
# print("\nTop Areas in KMeans Cluster 0:")
# display(df_sampled[df_sampled['KMeans_Cluster'] == 0]['Area'].value_counts().head())
#
# print("\nTop Items in KMeans Cluster 0:")
# display(df_sampled[df_sampled['KMeans_Cluster'] == 0]['Item'].value_counts().head())
```
